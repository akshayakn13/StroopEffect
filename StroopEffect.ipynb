# Stroop Effect

## Background Information

In a Stroop task, participants are presented with a list of words, with each word displayed in a color of ink. The participantâ€™s task is to say out loud the color of the ink in which the word is printed. The task has two conditions: a **congruent words condition**, and an **incongruent words condition**. In the **congruent words** condition, the words being displayed are color words whose names match the colors in which they are printed: for example RED, BLUE. In the **incongruent words condition**, the words displayed are color words whose names do not match the colors in which they are printed: for example PURPLE, ORANGE. In each case, we measure the time it takes to name the ink colors in equally-sized lists. Each participant will go through and record a time from each condition.

### Now, perform the statistical test and report your results. What is your confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

**Confidence Interval** = 95%



**alpha level** = 0.025 for two tailed t-test



**Degree of freedom** = n1 + n2 - 2 => 24+24 - 2

**Pooled Variance(Sp^2)** = (SScongruent + SSincongruent)/(degree of freedom)


where , SS is the sum of square

```python
## Calculating sum of squares
mean_congruent = df["Congruent"].mean()
ss_cong =0 
for i in df["Congruent"]:
    ss_cong += (i-mean_congruent)**2
    
print(mean_congruent)
print(ss_cong)
```

```python
## Calculating sum of squares
mean_incongruent = df["Incongruent"].mean()
ss_incong =0 
for i in df["Incongruent"]:
    ss_incong += (i-mean_incongruent)**2
    
print(mean_incongruent)
print(ss_incong)
```

```python
## Calculating pooled variance 
n1 = df["Congruent"].count()
n2 = df["Incongruent"].count()

print(f"No of sample in congruent = {n1} , incongruent = {n2}")
pooled_var = (ss_cong+ss_incong)/(n1 + n2 -2)

print(pooled_var)
```

**Standard Error** = root((2sp^2)/(n1+n2))

```python
## Calculating standard error 
SE = math.sqrt((2*(pooled_var**2))/(n1+n2))
print(SE)
```

**t-statistics =** difference(mean of incongruent, mean of congruent)/(standard error)

```python
#calculating t-statistics 

t= (mean_incongruent - mean_congruent)/SE

print(f"t-statistics is +-{t}")
```

**t-critical value for a two-tailed t-test with degree of freedom = 46 is +- 1.684**


Since p-value lies in the critical region we can say that t-value is statistically significant at the alpha level = 0.05 and hence **We reject the null hypothesis that means we accept the alternative hypothesis**
